# Data Preprocessing

```python
match_df['team1'].unique()
```

```python
# name of those 8 teams those are playing currently
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
# Now Delhi Daredevils and Deccan Chargers teams are renamed as Delhi Capitals and Sunrisers Hyderabad respectively
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df['dl_applied'].value_counts()
# 15 times duckwort luice is applied in the ipl
```

```python
match_df = match_df[match_df['dl_applied'] == 0]
# we will consider only those matchs when duckworth luice is not applied)
```

```python
match_df.head(2)
```

```python
# fetching only useful columns from match_df dataframe
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
delivery.head(2)
```

```python
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
# considering only second inning matches because the team that tries to 
# chase the winning score plays second inning
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df.sample(3)
```

```python
# finding the score after each ball for every match
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
# how many runs are needed to win after the current score
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
# how many balls are left
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
# now we will find out that how many wickets are left
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
```

```python
delivery_df['player_dismissed'].unique()
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
```

```python
delivery_df.head(4)
```

```python
# finding the cumulative sum of wickets for each match
wickets=delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
```

```python
# how many wickets are left
delivery_df['wickets'] = 10 - wickets
```

```python
delivery_df.head()
```

```python
#  current runrate=current score per over
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
# required runrate
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df.head(2)
```

```python
delivery_df.loc[(delivery_df["winner"]==delivery_df["batting_team"])]
```

```python
def result(dataframe):
    return 1 if dataframe['batting_team'] == dataframe['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
# taking only important columns for prediction
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df.head(2)
```

```python
# shuffling the final dataset
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.head(3)
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.describe()
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
